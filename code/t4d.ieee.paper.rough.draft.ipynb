# III. Methods

## Overall Workflow

- The workflow is staged to match physical intuition and keep interpretation manageable.
- Stage 1 establishes station context, baselines, and cohort quality screens.
- Stage 2 models climate-to-water temperature response.
- Stage 3 models downstream chemistry response from annual station-year conditions.
- Stage 4 projects future station behavior under warming scenarios.

## Why A Staged Method Was Used

- It is easier to explain than a single opaque end-to-end model.
- It lets each target use features that make sense for that level of the process.
- It makes debugging and sanity checks more transparent.
- It aligns with the project notebook phases already developed.


## III-A. Regime Context, Baselines, and Cohorts

- Build station baselines from historical observations.
- Retain only years with enough coverage to support annual summaries.
- Use station descriptors such as annual means, seasonal amplitude, timing markers, and coverage statistics.
- Use these descriptors both for context and as static predictors in later stages.
- Split stations or station-years into usable and flagged cohorts.

### Why This Step Matters

- Baselines anchor later delta-style predictions.
- Coverage filters reduce the chance that thin data define the station norm.
- Cohorts help separate stronger evidence from weaker evidence.

# V. Discussion and Limitations

## Interpretation Points

- The staged model structure is physically intuitive but still statistically approximate.
- Forecast skill likely differs by target because estuarine processes vary in predictability.
- Station-level models preserve useful heterogeneity that would be lost in a single aggregate model.

## Key Limitations

- Historical relationships may shift under future climate conditions.
- Missingness and sparse coverage can create selection effects.
- Annual summaries may miss short-duration extreme events.
- Some outcomes may require process-based covariates not included here.
- The workflow emphasizes interpretability and tractability over maximal complexity.

## Caveats About Forecast Use

- Treat outputs as scenario-guided estimates, not deterministic truth.
- Low-error average behavior does not guarantee correct extreme-event prediction.
- Strong local fit does not prove causal structure.
- Model confidence should be lower for thinly observed stations or targets.


## Optional Classification Extension

- If you later define a binary risk target, such as hypoxic versus non-hypoxic season, then ROC curves could become useful.
- ROC curves show how well a model separates two classes across many decision thresholds.
- They are not usually the first plot for continuous regression targets like temperature or oxygen mean change.
- For the current paper, observed-versus-predicted scatterplots are probably more natural for most core results.
- If a binary thresholded risk outcome is added, then include ROC and precision-recall curves in a short extension subsection.

## Simple ROC Note For Later

- X-axis:
  false positive rate
- Y-axis:
  true positive rate
- Better curves bend toward the upper-left corner.
- Area under the curve summarizes ranking ability across thresholds.
- ROC is best when the problem is genuinely classification, not just thresholding a regression output after the fact without justification.
